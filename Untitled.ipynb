```python
def visualize_cs_mask(cs_mask):
    image = np.zeros((*cs_mask.shape, 3))
    image[cs_mask==0, :] = (0., 1., 0.)
    image[cs_mask==1, :] = (1., 0., 0.)
    plt.imshow(image)
    plt.show()

example_cs_mask = np.load("prepared_data/unified_masks/classification_masks/0000.npy")
visualize_cs_mask(example_cs_mask)
```

```python
def visualize_ps_mask(ps_mask):
    ps_mask = np.exp(ps_mask) / np.exp(ps_mask).sum()
    print((ps_mask != 0).sum())
    plt.imshow(ps_mask)
    plt.show()

example_ps_mask = np.load("prepared_data/unified_masks/positive_sampling_segmentation_masks/0000.npy")
visualize_ps_mask(example_ps_mask)
```

```python
def visualize_ns_mask(ns_mask):
    ns_mask = np.exp(ns_mask) / np.exp(ns_mask).sum()
    
    plt.imshow(ns_mask)
    plt.show()

example_ns_mask = np.load("prepared_data/unified_masks/negative_sampling_segmentation_masks/0000.npy")
visualize_ps_mask(example_ns_mask)
```
